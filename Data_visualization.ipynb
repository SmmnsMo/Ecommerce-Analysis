```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
oct_19 = pd.read_csv('oct_19_clean.csv', index_col = 0)
nov_19 = pd.read_csv('nov_19_clean.csv', index_col = 0)
dec_19 = pd.read_csv('dec_19_clean.csv', index_col = 0)
jan_20 = pd.read_csv('jan_20_clean.csv', index_col = 0)
feb_20 = pd.read_csv('feb_20_clean.csv', index_col = 0)
```

```python
yticks = []
for i in range(0, 2000000, 250000):
    yticks.append(i)
```

```python
fig, ax = plt.subplots()
b = plt.bar(x = ['View', 'Cart', 'Remove From Cart', 'Purchase'], height = oct_19['event_type'].value_counts().values)
ax.yaxis.get_major_formatter().set_scientific(False)
ax.yaxis.get_major_formatter().set_useOffset(False)
ax.bar_label(ax.containers[0], labels=[f'{x.get_height():.0f}' for x in ax.containers[0]])
plt.xlabel('Event Type')
plt.ylabel('Count of Events')
plt.yticks(yticks)
plt.title('October 2019 Event Funnel')
plt.show()
```
